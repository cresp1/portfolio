### 4. Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans 
```

```python
comp_brasil = pd.read_csv('COMPUTACAO_BRASIL.csv', sep=';', na_values=' ')
comp_brasil.fillna(0, inplace=True)
```

```python
# criando coluna de inscritos por vaga (disputa)
comp_brasil['INSCRITOS_POR_VAGA'] = round(comp_brasil['QT_INSCRITO_TOTAL'] / comp_brasil['QT_VAGA_TOTAL']) 
comp_brasil.head()
```

```python
# devido ao último cálculo, 11 registros ficaram índice de disputa (inscritos por vaga) nulo
comp_brasil.dropna(inplace=True)
```

Para analisar com maior distinção dos dados, vamos discriminar as universidades entre públicas e privadas. Assim, visto que a PUC-SP é uma universidade privada, trabalharemos com essas.

```python
colunas = ['NO_IES', 'QT_MATRICULA_TOTAL', 'QT_CONCLUINTE_TOTAL', 'QT_INGRESSO_TOTAL', 
           'QT_VAGA_TOTAL', 'QT_INSCRITO_TOTAL','INSCRITOS_POR_VAGA']

# universidades públicas
comp_brasil_pub = comp_brasil.query('IN_GRATUITO == 1').filter(items=colunas)

# universidades privadas
comp_brasil_priv = comp_brasil.query('IN_GRATUITO == 0').filter(items=colunas)
```

```python
# comp_brasil_pub.head(3)
```

```python
comp_brasil_priv.head(3)
```

```python
X = comp_brasil_priv.drop(columns='NO_IES')
```

```python
scaler = StandardScaler()
priv = scaler.fit_transform(X)
```

```python
# método elbow para encontrar melhor k
wcss = []
 
for i in range(1, 21):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(priv)
#     print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_) 
plt.figure(figsize=(10, 4))
plt.plot(range(1, 21), wcss)
plt.title('O Método Elbow')
plt.xlabel('Número de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
# modelo KMeans
km_priv = KMeans(n_clusters=5, n_init= 16, random_state=65, algorithm='full')
km_priv.fit(priv)
```

```python
# rótulos
# km_priv.labels_
```

```python
# nova coluna com os rótulos
comp_brasil_priv['Grupo'] = km_priv.labels_
```

```python
grupo_puc = comp_brasil_priv[comp_brasil_priv.NO_IES == 'PONTIFÍCIA UNIVERSIDADE CATÓLICA DE SÃO PAULO']['Grupo']
grupo_puc = int(grupo_puc)
print(f'A PUC-SP se encontra no grupo {grupo_puc}')
```

```python
# quantas universidades há em cada grupo
comp_brasil_priv.groupby('Grupo')[['NO_IES']].count().style.bar()
```

```python
comp_brasil_priv.groupby('Grupo').mean().style.highlight_max()
```

**Observações:** O grupo no qual a PUC-SP se encontra conta com a maior média de índice de disputa (quantidade de inscritos por vaga)

O grupo também conta com a terceira maior média de matrículas e de concluintes no ano de 2019.

### 5. Classificador 

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
# divisão dos dados em treino e teste
y = comp_brasil_priv['Grupo']
X = comp_brasil_priv.drop(['Grupo','NO_IES'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
# ajustando o modelo aos dados
clf = svm.SVC(C=1.0)
clf.fit(X_train,y_train)
```

```python
# score geral do modelo
y_pred = clf.predict(X_test)
clf.score(X_test,y_test)
```

### 5.1 Métricas de avaliação

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
```

### 5.2 Matriz de confusão

```python
# matriz de confusão
cf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 5))
sns.heatmap(cf_matrix, annot=True, cmap='Blues');
```

### 5.4 Accuracy score

```python
accuracy_score(y_test, y_pred)
```

### 5.5 F1 score

```python
f1_score(y_test, y_pred, average='weighted')
```

### 5.6 Precision score

```python
precision_score(y_test, y_pred, average='weighted', zero_division=0)
```

### 5.7 Recall score

```python
recall_score(y_test, y_pred, average='weighted')
```
